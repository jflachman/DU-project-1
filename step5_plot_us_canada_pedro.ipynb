```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px
import fastparquet
```

```python
US_CA_Ground_Stations_n_locations = pd.read_parquet('result_files/us_ca_weather_bins.parquet.gzip', engine="fastparquet") 
US_CA_Ground_Stations_n_locations.dtypes
```

```python
US_CA_Ground_Stations_n_locations.columns
```

### Plotly Maps

- Colors.sequential
    - Sunset
    - Agsunset
    -Jet  (blue to red)
-colors.diverging
    - 

```python
plotly_html_path = "../../jflachman.github.io/plotly/"
```

```python

fig90 = px.scatter_mapbox(US_CA_Ground_Stations_n_locations, lat="Lat", lon="Lon", 
                         hover_name="Name", 
                         hover_data=["State", "Elevation", 'Days<0', 'Days<10',
       'Days<20', 'Days<32', 'Days<40', 'Days<55', 'Days<60', 'Days<65','Days>60', 'Days>65', 'Days>70', 'Days>75', 'Days>80',
       'Days>85', 'Days>90', 'Days>100', 'Days>110', 'Days-75-95', 'Days-25-55', 'Days-60-80' ],
                        color="Days>90", 
                        color_continuous_scale=px.colors.sequential.Jet, 
                        range_color=[0,100],
                        zoom=3, width=900, height=600)
fig90.update_layout(mapbox_style="open-street-map")
fig90.update_layout(margin={"r":10,"t":10,"l":10,"b":10})

# fig90.write_html(plotly_html_path + "greater90.html")
fig90.show()
```
